```python
import pandas as pd
df = pd.read_csv('clean.txt')
df.head()
```

```python
df = df.fillna(0)
df.isnull().sum()
```

```python
# drop players with <16 games
vc = (df['player'].value_counts() >= 16).to_dict()
enouf = [p for p in vc if vc[p]]
df = df[df['player'].isin(enouf)]
df.head()
```

```python
df_data = df[['net%', 'netatt', 'ny', 'ny/a', 'car', 'ry', 'ypc', 'tds', 'tos', 'player']]
df_data.head()
```

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df_data, test_size=0.2, random_state=143, stratify=df_data['player'])

y_col = 'player'
X_cols = df_train.columns.drop(y_col)

X_train = df_train[X_cols]
X_test = df_test[X_cols]
y_train = df_train[y_col]
y_test = df_test[y_col]
```

```python
# Dummy majority baseline
from sklearn.dummy import DummyClassifier
model_dum = DummyClassifier(strategy='most_frequent')
model_dum.fit(X_train, y_train)
model_dum.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model_rfc = RandomForestClassifier(
    n_estimators=100,
    max_depth=8,
    min_samples_split=20,
    n_jobs=-1,
    random_state=143
)
model_rfc.fit(X_train, y_train)
model_rfc.score(X_test, y_test)
```
