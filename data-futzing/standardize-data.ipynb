```python
import pandas as pd
df0 = pd.read_csv('clean.txt')

def quickpeek(df):
    print(df.shape)
    print(f"{df['player'].nunique()} QBs")
    display(df.head())

quickpeek(df0)
```

```python
# standardize playing stats by league year
# (justification for this can be found in `change-over-year.ipynb`)
stats = ['cmp%', 'attempts', 'ny/a', 'carries', 'ypc', 'td:to']
years = df0['season'].unique()

df1 = df0.copy()
for s in stats:
    for yr in years:
        onestatoneyear = df0[df0['season'] == yr][s]
        m = onestatoneyear.mean()
        sd = onestatoneyear.std()
        df1.loc[df1['season'] == yr, s] = (onestatoneyear - m) / sd

quickpeek(df1)
```

```python
# standardize metadata
# (justification: so that everything has the same "unit")
nonstats = ['season', 'quint', 'home']

df2 = df1.copy()
for col in nonstats:
    df2[col] = (df1[col] - df1[col].mean()) / df1[col].std()

quickpeek(df2)
```

```python
df2.to_csv('standard.txt', index=False)
```
