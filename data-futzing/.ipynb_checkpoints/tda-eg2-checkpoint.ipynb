```python
import pandas as pd
df = pd.read_csv('clean-pass.txt')
df.head()
```

```python
df_tda = df[['net%', 'netatt', 'ny', 'ny/a']]
df_tda.head()
```

```python
import gudhi as gd
skel = gd.RipsComplex(points=df_tda.values.tolist(), max_edge_length=0.01)
simp = skel.create_simplex_tree(max_dimension=3)
simp.num_simplices()
```
