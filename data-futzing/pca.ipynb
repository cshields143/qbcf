```python
import pandas as pd
df0 = pd.read_csv('standard.txt')

print(df0.shape)
df0.head()
```

```python
# use PCA to reduce the number of features
from sklearn.decomposition import PCA
reducer = PCA(n_components=3, random_state=143)
obfus = reducer.fit_transform(df0[df0.columns.drop('player')])
df1 = pd.DataFrame(obfus, columns=['pca0', 'pca1', 'pca2'])
df1['player'] = df0['player']
df1 = df1[['player', 'pca0', 'pca1', 'pca2']]

df1.head()
```

```python
df1.to_csv('pca.txt', index=False)
```
